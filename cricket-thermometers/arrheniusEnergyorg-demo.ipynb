# Model 1: Reaction rates at two temperatures



The relationship between a reaction's rate at two temperatures depends on $E_a$ the activation energy:

$$ \ln \dfrac{r_2}{r_1}=\dfrac{E_{\textrm a}}{R}\left(\dfrac{1}{T_1}-\dfrac{1}{T_2}\right) \nonumber $$

$r_1$ and $r_2$ correspond to the rate at temperatures $T_1$ and $T_2$, respectively.

The thermal energy available at room temperature is given by $kT$ (per molecule) or $RT$ (per mole): 4.1e-21 J/molec = 2.5 kJ/mol.

# Model 3: The Arrhenius equation



The temperature dependence of rates can be modeled with the
<a href="https://en.wikipedia.org/wiki/Arrhenius_equation">Arrhenius
Equation</a>:

$$ k = Ae^{-E_{a}/RT} \qquad\qquad\quad (1) $$

or, in linear form:

$$ \ln k = \frac{-E_a}{R} \frac{1}{T} + \ln(A)\qquad\qquad (2) $$

where $A$=pre-exponential factor, $E_a$=activation energy, and $R$=gas
constant.

# Fitting data with curve\_fit




## Worked Example: Fitting data with scipy.stats.linregress



Signature for `linregress()`


Use `linregress()` to perform a regression on the data.


```python
x = np. array([ 0.,  1.4,  2.8,  4.3,  5.77, 7.2,  8.6, 10.1, 11.5, 13.])
y = np. array([ 0.29, 3.2, 2.6, 4.3, 5.5, 7.3, 8.7, 11.2, 14.4, 15.1])
reg = linregress(x, y)
print(reg)
```

Check the fit by generating y values based on the regression and plotting both the fit and the original data.


```python
y_fit = x * reg.slope + reg.intercept 
plt.plot(x, y_fit, label='best fit')
plt.plot(x, y, 'o', label='data')
plt.legend()
```

# Dissociation energies of CFCs



Bromine and chlorine containing compounds are problematic because they catalyze the decomposition of ozone.


```python
h2kj_mol = 2625.50   # Hartree to kJ/mol
h2kcal_mol = 627.509 # Hartree to kcal/mol

Eopt = -499.0931526167 # hartree; optimized energy of CH3Cl from webmo
Rs, Es = np.genfromtxt("ch3cl-scan1237.csv", skip_header=1, delimiter=',', unpack=True)
# fig, ax = plt.subplots()
# ax.plot(Rs, Es, '-o' )
BEch3cl = (Es[-1] - Eopt) * h2kj_mol # kJ/mol
BEch3cl
```

CH<sub>3</sub>Br bond energy.


```python
E_max = -2611.444783 # hartree; E at 10AA
BEch3br = (E_max + 2611.717972610017) * h2kj_mol
BEch3br
```


# Plotting and FFT of wav



Scipy can plot wav files.


```python
from scipy.io.wavfile import read
import matplotlib.pyplot as plt

# read audio samples
sample_rate12C, data_12C = read("./cricket-data/585ssj.wav")
sample_rate30C, data_30C = read("./cricket-data/585ssh.wav")
length = data_12C.shape[0]/sample_rate12C # seconds
time12C = np.linspace(0, length, data_12C.shape[0])
time30C = np.linspace(0, length, data_30C.shape[0])

fig, ax = plt.subplots(2,1, figsize=(10,5))
ax[0].plot(time12C, data_12C, lw=0.25)
ax[1].plot(time30C, data_30C, lw=0.25);
```
